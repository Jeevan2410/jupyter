```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
data = pd.read_csv('Social_Media.csv')
data.head()  # Preview the dataset
```

Data Cleaning:


```python
# Check for missing values
print(data.isnull().sum())

# If any missing values are found, we can handle them (for example, by removing rows with missing values)
data = data.dropna()
```

 Analyzing the Data:

Research Question 1: How does age affect productivity loss due to social media usage?

```python
# Calculate the correlation between age and productivity loss
age_vs_productivity_corr = data['Age'].corr(data['ProductivityLoss'])

# Plot the relationship
plt.figure(figsize=(8, 6))
plt.scatter(data['Age'], data['ProductivityLoss'], alpha=0.5)
plt.title(f"Impact of Age on Productivity Loss (Correlation: {age_vs_productivity_corr:.2f})")
plt.xlabel("Age")
plt.ylabel("Productivity Loss")
plt.grid(True)
plt.show()
```

Research Question 2: Which social media platforms contribute the most to productivity loss?

```python
# Group the data by platform and calculate average productivity loss
platform_vs_productivity = data.groupby('Platform')['ProductivityLoss'].mean().sort_values(ascending=False)

# Plot the result
plt.figure(figsize=(10, 6))
platform_vs_productivity.plot(kind='bar', color='skyblue')
plt.title("Average Productivity Loss by Social Media Platform")
plt.xlabel("Platform")
plt.ylabel("Average Productivity Loss")
plt.xticks(rotation=45)
plt.grid(True, axis='y')
plt.show()
```

Research Question 3: Do males and females experience productivity loss differently?

```python
# Group the data by gender and calculate average productivity loss
gender_vs_productivity = data.groupby('Gender')['ProductivityLoss'].mean()

# Plot the result
plt.figure(figsize=(8, 6))
gender_vs_productivity.plot(kind='bar', color=['lightcoral', 'lightblue'])
plt.title("Average Productivity Loss by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Productivity Loss")
plt.xticks(rotation=0)
plt.grid(True, axis='y')
plt.show()
```

Research Question 4: How does addiction to social media correlate with reduced productivity?

```python
# Calculate the correlation between addiction level and productivity loss
addiction_vs_productivity_corr = data['Addiction Level'].corr(data['ProductivityLoss'])

# Plot the relationship
plt.figure(figsize=(8, 6))
plt.scatter(data['Addiction Level'], data['ProductivityLoss'], alpha=0.5)
plt.title(f"Addiction Level vs Productivity Loss (Correlation: {addiction_vs_productivity_corr:.2f})")
plt.xlabel("Addiction Level")
plt.ylabel("Productivity Loss")
plt.grid(True)
plt.show()
```

Research Question 5: What is the role of self-control in mitigating the negative effects of social media usage?

```python
# Calculate the correlation between self-control and productivity loss
self_control_vs_productivity_corr = data['Self Control'].corr(data['ProductivityLoss'])

# Plot the relationship
plt.figure(figsize=(8, 6))
plt.scatter(data['Self Control'], data['ProductivityLoss'], alpha=0.5)
plt.title(f"Self-Control vs Productivity Loss (Correlation: {self_control_vs_productivity_corr:.2f})")
plt.xlabel("Self-Control")
plt.ylabel("Productivity Loss")
plt.grid(True)
plt.show()
```

Research Question 6: Does user satisfaction influence the degree of productivity loss?

```python
# Calculate the correlation between satisfaction and productivity loss
satisfaction_vs_productivity_corr = data['Satisfaction'].corr(data['ProductivityLoss'])

# Plot the relationship
plt.figure(figsize=(8, 6))
plt.scatter(data['Satisfaction'], data['ProductivityLoss'], alpha=0.5)
plt.title(f"Satisfaction vs Productivity Loss (Correlation: {satisfaction_vs_productivity_corr:.2f})")
plt.xlabel("Satisfaction")
plt.ylabel("Productivity Loss")
plt.grid(True)
plt.show()
```

Key Correlations Between Demographic and Behavioral Factors:

```python
# Calculate correlation matrix for key variables
correlation_matrix = data[['Age', 'ProductivityLoss', 'Addiction Level', 'Self Control', 'Satisfaction']].corr()

# Plot the correlation matrix
plt.figure(figsize=(8, 6))
plt.matshow(correlation_matrix, cmap='coolwarm', fignum=1)
plt.colorbar()
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=45)
plt.yticks(range(len(correlation_matrix.index)), correlation_matrix.index)
plt.title("Correlation Matrix of Demographic and Behavioral Factors", pad=20)
plt.show()
```

Conclusion:
From the analysis, we derive the following insights:

Age has a minimal impact on productivity loss due to social media usage.

YouTube is the platform with the highest productivity loss, followed by TikTok, Instagram, and Facebook.

Gender differences in productivity loss are minimal.

There is a positive correlation between addiction level and productivity loss.

Self-control plays a key role in reducing productivity loss, as it is negatively correlated with productivity loss.

Higher user satisfaction is slightly correlated with lower productivity loss, indicating better management of time or user perception.
